```python
""" Possible metrics affecting threshold:
- Number of runs
- Number of source edits
- Time spent
- Total lines changed (total line diff)

If these metrics are high, we could decrease the threshold? I.e. there is a greater chance of classifying as struggling

"""

num_data_points = len(df)
num_source_edits = df.SourceEdit.sum()
num_runs = df.TotalRuns.sum()
num_minutes_spent = round(df.Relative_time.iloc[-1].seconds / 60)

total_line_diff = df.filter(regex='^Line_diff\d{1,}$').sum().sum()
```

```python
def calculate_em_threshold(series):
    return series.expanding().mean()
```

```python
def pct_diff_from_threshold(threshold, series):
    diff = series - threshold
    return diff / threshold
```

```python
threshold = calculate_em_threshold(df.scaled_max_smoothed)
```

```python
pct = pct_diff_from_threshold(threshold, df.scaled_max_smoothed)
```
